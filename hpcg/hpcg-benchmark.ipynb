```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("performance.csv")
```

```python
df.dtypes
```

```python
df.columns
```

```python
for arch in df.ARCH.unique():
    sorted_df = df[df.ARCH == arch].sort_values(by="NX")
    plt.plot(sorted_df['NX'],sorted_df['GFLOP'],label=f"{arch}")
plt.legend()
plt.xlabel("Nx")
plt.ylabel("GFLOP/s")
plt.title("Performance")
plt.savefig("Archxperf.png")
plt.show()
```
